```
import matplotlib.pyplot as plt
```

```
# BZ TREE GRAPH - THREADS VS THROUGHPUT

threads = [1, 8, 16, 32, 64]
bzThroughputReadOnly = [953072, 6172321, 7980997, 9056828, 9386497]
bzThroughput10Write = [718435, 4656001, 6449443, 7434534, 6688198]
bzThroughput50Write = [354642, 2484297, 3716887, 3399154, 1390102]
plt.figure(figsize=(10,8))
bzReadOnlyLine, = plt.plot(threads, bzThroughputReadOnly, color='g', label = 'Read Only')
bzWrite10Line, = plt.plot(threads, bzThroughput10Write, color='orange', label = '10% Write')
bzWrite50Line, = plt.plot(threads, bzThroughput50Write, color='black', label = '50% Write')
plt.xlabel('Threads')
plt.ylabel('Throughput in ops/sec')
plt.title('bzTree Benchmarking (Threads v/s Throughput)')
plt.legend(handles=[bzReadOnlyLine, bzWrite10Line, bzWrite50Line])
```

```
# BZ TREE - MEDIAN LATENCY / P99 LATENCY v/s THREADS

threads = [1, 8, 16, 32, 64]
bzMedianLatencyReadOnly = [679, 838, 1138, 1315, 1639]
bzMedianLatency10Write = [701, 867, 1200, 1352, 1349]
bzMedianLatency50Write = [2950, 3281, 3146, 3109, 3044]
bz99LatencyReadOnly = [1085, 1319, 1864, 1965, 1953]
bz99Latency10Write = [3544, 4077, 5052, 5188, 5184]
bz99Latency50Write = [21086, 24313, 32311, 33427, 32618]
plt.figure(figsize=(10,8))
bzMedianReadOnlyLine, = plt.plot(threads, bzMedianLatencyReadOnly, color = 'g', label = 'Median Latency Read Only')
bzMedian10WriteLine, = plt.plot(threads, bzMedianLatency10Write, color = 'orange', label = 'Median Latency 10% Write')
bzMedian50WriteLine, = plt.plot(threads, bzMedianLatency50Write, color = 'black', label = 'Median Latency 50% Write')
bz99LatencyReadOnlyLine, = plt.plot(threads, bz99LatencyReadOnly, color= 'red', label = '99 Percentile Latency Read Only')
bz99Latency10WriteLine, = plt.plot(threads, bz99Latency10Write, color= 'blue', label = '99 Percentile Latency 10% Write')
bz99Latency50WriteLine, = plt.plot(threads, bz99Latency50Write, color= 'yellow', label = '99 Percentile Latency 50% Write')
plt.xlabel('Threads')
plt.ylabel('Latency in milliseconds')
plt.title('Median & 99 Percentile Latencies for bzTree')
plt.legend(handles = [bzMedianReadOnlyLine, bzMedian10WriteLine, bzMedian50WriteLine, bz99LatencyReadOnlyLine, bz99Latency10WriteLine, bz99Latency50WriteLine])
```
